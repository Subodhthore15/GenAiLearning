## Huggingface Endpoint

```python
from langchain_huggingface import HuggingFaceEndpoint
repo_id = "mistralai/Mistral-7B-Instruct-v0.3"
# repo_id = "gpt-omni/mini-omni2"
repo_id_text_image = "Helsinki-NLP/opus-mt-zh-en"
```

```python
# llm=HuggingFaceEndpoint(repo_id=repo_id,
#                          max_new_tokens=500,
#                            temperature=0.2,
#                            huggingfacehub_api_token= sec_data      )
llm=HuggingFaceEndpoint(repo_id=repo_id,
                         model_kwargs = {"temparature":0.4, "token":sec_data, "max_length":500}     )
```

```python
llm.invoke("what is machine learning?")
```

```python
from langchain import PromptTemplate, LLMChain

template = """
for the given Question :{question}.
answer it in step by step .
 """
prompt = PromptTemplate(template = template, input_variables = ['question'])
```

```python
prompt
```

```python
chain=LLMChain(llm = llm, prompt= prompt)
chain.invoke("who won the cricket world cup in 2011?")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
hf.embed_query("what is the machine learning?")
```

```python
from huggingface_hub import InferenceApi

# Set up the API with your Hugging Face token
# inference = InferenceApi(repo_id="stabilityai/stable-diffusion-2-1", token=sec_data)
inference = InferenceApi(repo_id="black-forest-labs/FLUX.1-dev", token=sec_data)



# Define the text prompt
prompt = "hrithik Roshan doing sword cutting fightingg scene"

# Generate the image
response = inference(inputs=prompt)

# Check if the response is an image and save it
if isinstance(response, dict) and 'error' in response:
    print(f"Error: {response['error']}")
else:
    # Save the generated image directly using the PIL Image object's save method
    response.save("generated_image.png")
    print("Image saved as 'generated_image.png'")

```
