### Prompt Template for Text Summarization

```python
from langchain.chains import LLMChain  # or prompt | LLM  means combine prompt with LLM.
from langchain import PromptTemplate 

genericPromptTemplate = """ 

  Write a summary of the following speech:
  Translate the precise summary only in {language} language. Don't give in English
  Speech: \n {speech}. 
  
"""


prompt=PromptTemplate(template = genericPromptTemplate , input_variables=["speech","language"])

#prompt.format(speech = speech, language = "french")
```

```python
chain=LLMChain(llm = llm, prompt = prompt)
#chain = prompt|llm
```

```python
response=chain.invoke({"speech" : speech, "language": "Marathi"})
```

## Map reduce techinique


```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("attention.pdf")
pages = loader.load_and_split()
```

```python
pages
```

```python
# make chunks of document
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size = 2000, chunk_overlap = 150)
splittedDocs=splitter.split_documents(pages)
```

```python
splittedDocs
```

```python
len(splittedDocs)
```

```python
chunks_prompts = """ 
    Please summarize the below speech:
    Speech: {text}
"""

map_promptTemplate = PromptTemplate(template = chunks_prompts , input_variable = ["text"])
```

```python
map_promptTemplate
```

```python
final_prompt = """ 
 Here provided all the summaries of entire documents with these important points.
  go and add the Precise Title, and provide the final summary. 
 \n {text}. 
 
"""
final_prompt = PromptTemplate(template = final_prompt , input_variables =["text"] )
```

```python
final_prompt
```

```python
summary_chain = load_summarize_chain(llm = llm, 
                                     chain_type="map_reduce",
                                       map_prompt= map_promptTemplate, 
                                       combine_prompt = final_prompt, verbose = True)
```

```python
summary_chain.invoke(splittedDocs)
```

## Refine 

```python
chain = load_summarize_chain(
    llm = llm,
    chain_type= "refine",
    verbose = True
)

chain.run(splittedDocs)
```
